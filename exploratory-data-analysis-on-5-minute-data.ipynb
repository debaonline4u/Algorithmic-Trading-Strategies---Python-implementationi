```python
# Import necessary libraries

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os

from matplotlib import pyplot as plt    # for plotting purposes
```

```python
# List of files availeble in the current directory
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
# Here clarify about the data is loaded or not
filename = 'NIFTY 50 - 5minute data.csv'
print(filename)
data_path = os.path.join(dirname, filename)
print('Data path: ', data_path)
```

```python
# Read dataset into dataframe
dataset = pd.read_csv(data_path)
print(dataset.head())
```

### Since this data belongs to index category, so volume data is not available

```python
plt.figure(figsize = (12, 6))
plt.title('Nifty 50 data')
plt.rcParams.update({'font.size': 14})
plt.xlabel('Time in 5 minutes')
plt.ylabel('Price')
plt.plot(dataset['close'])
```

* Categorical data refers to a data type that can be stored and identified based on the names or labels given to them.
* Numerical data refers to the data that is in the form of numbers, and not in any language or descriptive form.
